```python
'''
What is a Tensor?
At its core, a tensor is a mathematical object that generalizes the concept of scalars, vectors, and matrices to higher dimensions.
In the context of data science, tensors are multi-dimensional arrays of numbers that represent complex data.
They are the fundamental data structures used in machine learning and deep learning frameworks like TensorFlow and PyTorch.
'''
```

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
v = torch.tensor([1,2,3,4,5,6])
print(v.dtype)
```

```python
f = torch.FloatTensor([1,2,3,4,5,6])
print(f.dtype)
```

```python
v.view(6,1) # returns 6 rows and 1 column
```

```python
v.view(3,2) # replace 2 with -1 and o/p is still the same
```

```python
# convert numpy ar to tensor
a = np.array([1,2,3,4,5])
tensor_cnv = torch.from_numpy(a)
print(tensor_cnv)
print(tensor_cnv.type())
```

```python
# convert tensor to numpy arr
numpy_cnv = tensor_cnv.numpy()
print(numpy_cnv)
```

```python
# Vector operations
t_one = torch.tensor([1,2,3])
t_two = torch.tensor([1,2,3])
t_one + t_two
```

```python
t_one = torch.tensor([1,2,3])
t_two = torch.tensor([1,2,3])
t_one * t_two
```

```python
t_one = torch.tensor([1,2,3])
t_two = torch.tensor([1,2,3])
t_one * 5
```

```python
t_one = torch.tensor([1,2,3])
t_two = torch.tensor([1,2,3])
dot_product = torch.dot(t_one,t_two)
print(dot_product)
```
