```python
import torch
```

```python
one_d = torch.arange(1,10) # arange() initilizes a 1D tensor
one_d
```

```python
two_d = one_d.view(3,3)
two_d
```

```python
two_d.dim() # returns the dimension of tensor --> a 2D matrix with 3x3 rows and columns
```

```python
# fetch elements from matrix
two_d[0,2] # first row(0th index) and 3rd column(2nd index)
```

```python
# 3D tensor
x = torch.arange(18).view(2,3,3)
x
```

```python
# SLICING 3D Tensors
x[0,1,1]
```

```python
x[1,:,:]
```

```python
# Matrix multiplication
mat_a = torch.tensor([0,3,5,5,5,2]).view(2,3)
mat_a
```

```python
mat_b = torch.tensor([3,4,3,-2,4,-2]).view(3,2)
mat_b
```

```python
# matrix multiplication --> torch.matmul()

torch.matmul(mat_a,mat_b)

mat_a @ mat_b # another way to do matrix mul
```

```python
# Derivatives

'''
If you make a tensor with:
 "requires_grad=True"

PyTorch will track all operations on that tensor.

Later, you can call:
"y.backward()"

and PyTorch will automatically compute dy/dx.
'''
```

```python
x = torch.tensor(2.0, requires_grad = True)
y = 9 * x ** 4 + 2 * x ** 3 + 3 * x ** 2 + 6 * x + 6
y.backward()
x.grad
```

```python
# Partial derivatives
x = torch.tensor(1.0, requires_grad = True)
z = torch.tensor(2.0, requires_grad = True)
y = x ** 2 + z ** 3
y.backward()
print(x.grad)
print(z.grad)
```
